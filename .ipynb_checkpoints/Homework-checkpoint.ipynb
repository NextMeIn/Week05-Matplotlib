```python
%matplotlib notebook
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

# File to Load (Remember to change these)
city_data_to_load = "data/city_data.csv"
ride_data_to_load = "data/ride_data.csv"

# Read the City and Ride Data
citydata = pd.read_csv(city_data_to_load)
ridedata = pd.read_csv(ride_data_to_load)
# Combine the data into a single dataset
data = pd.merge(ridedata,citydata,on='city',how="left")
# Display the data table for preview
data.head()
data.to_csv('data.csv')
```

```python
data = data.groupby(['type','city'],as_index=False).agg({'ride_id':'count','fare':'mean','driver_count':'max'})
data = data.rename(columns={'ride_id':'rides','fare':'avgfare'})
data['color'] = ""
```

```python
data.loc[data['type']=='Rural',['color']] = "khaki"
data.loc[data['type']=='Suburban',['color']] = "powderblue"
data.loc[data['type']=='Urban',['color']] = "lightsalmon"
```

```python
x_axis = data.loc[data['type']=='Urban']['rides']
y_axis = data.loc[data['type']=='Urban']['avgfare']
size = [x*10 for x in [data.loc[data['type']=='Urban']['driver_count']]]
type1 = plt.scatter(x_axis, y_axis, marker="o", edgecolors="black",s=size, alpha=1, color='lightsalmon', label='Urban')

x_axis = data.loc[data['type']=='Rural']['rides']
y_axis = data.loc[data['type']=='Rural']['avgfare']
size = [x*10 for x in [data.loc[data['type']=='Rural']['driver_count']]]
type2 = plt.scatter(x_axis, y_axis, marker="o", edgecolors="black",s=size, alpha=1, color='khaki', label='Rural')

x_axis = data.loc[data['type']=='Suburban']['rides']
y_axis = data.loc[data['type']=='Suburban']['avgfare']
size = [x*10 for x in [data.loc[data['type']=='Suburban']['driver_count']]]
type3 = plt.scatter(x_axis, y_axis, marker="o", edgecolors="black",s=size, alpha=1, color='powderblue', label='Suburban')

plt.title("Pyber Ride Sharing Data (2016)")
plt.xlabel("Total Number of Rides (Per City)")
plt.ylabel("Average Fate ($)")
plt.legend(handles=[type1, type2, type3], loc="upper right")
plt.grid()
plt.show()
```
