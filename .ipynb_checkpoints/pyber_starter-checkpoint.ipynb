```python
%matplotlib inline
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

# File to Load (Remember to change these)
city_data_to_load = "data/city_data.csv"
ride_data_to_load = "data/ride_data.csv"

# Read the City and Ride Data
citydata = pd.read_csv(city_data_to_load)
ridedata = pd.read_csv(ride_data_to_load)
# Combine the data into a single dataset
data = pd.merge(ridedata,citydata,on='city',how="left")
# Display the data table for preview
data.head()
```

```python
data['color'] = ""

data.loc[data['type']=='Rural',['color']] = "khaki"
data.loc[data['type']=='Suburban',['color']] = "powderblue"
data.loc[data['type']=='Urban',['color']] = "lightsalmon"
data.head()
```

```python
scatterdata = data.groupby(['type','city'],as_index=False).agg({'ride_id':'count','fare':'mean','driver_count':'max','color':'max'})
scatterdata = scatterdata.rename(columns={'ride_id':'rides','fare':'avgfare'})
```

```python
set(scatterdata['type'])
```

```python
legend, color = [], []

for x, y in scatterdata.iterrows():
    if y['type'] not in legend:
        legend.append(y['type'])
        color.append(y['color'])

legenddata=list(zip(legend,color))
legenddata
```

```python
fig = plt.figure(figsize=(10, 6))
ax = fig.add_subplot(111)

for x, y in scatterdata.iterrows():
    plt.scatter(y['rides'], y['avgfare'], marker="o", edgecolors="black", s=y['driver_count']*20, 
                alpha=1, color=y['color'], label=y['type']) 
    if y['type'] not in legend:
        legend.append(y['type'])
    if y['color'] not in color:
        color.append(y['color'])
    
plt.title("Pyber Ride Sharing Data (2016)")
plt.xlabel("Total Number of Rides (Per City)")
plt.ylabel("Average Fate ($)")
plt.grid()
plt.savefig('PyberRideSharingData')
plt.show()
```

## Bubble Plot of Ride Sharing Data

## Total Fares by City Type

```python
# Calculate Type Percents

# Build Pie Chart

# Save Figure
fig = plt.figure(figsize=(14, 6))
ax = fig.add_subplot(111)
piedata = data.groupby(['type'], as_index=False).agg({'fare':'sum','color':'max'})
labels = list(piedata['type'].values)
sizes = list(piedata['fare'].values)
colors = list(piedata['color'].values)

plt.pie(sizes, explode=(0,0,0.1), labels=labels, colors=colors,
        autopct="%1.1f%%", shadow=True, startangle=150)
plt.title("% of Total Fares by City Type")
plt.savefig('TotalFaresbyCityType.png')
plt.show()
```

## Total Rides by City Type

```python
# Calculate Ride Percents

# Build Pie Chart

# Save Figure
fig = plt.figure(figsize=(14, 6))
ax = fig.add_subplot(111)
piedata = data.groupby(['type'], as_index=False).agg({'ride_id':'count','color':'max'})
labels = list(piedata['type'].values)
sizes = list(piedata['ride_id'].values)
colors = list(piedata['color'].values)

plt.pie(sizes, explode=(0,0,0.1), labels=labels, colors=colors,
        autopct="%1.1f%%", shadow=True, startangle=150)
plt.title("% of Total Rides by City Type")
plt.savefig('TotalRidesbyCityType.png')
```

```python
# Show Figure
plt.show()
```
